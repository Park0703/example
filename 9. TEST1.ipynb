```python
import pandas as pd
data = {'product':['사과','딸기','수박'],
        'price':[1800,1500,3000],
        'qty':[24,38,13]}
ds1 = pd.DataFrame(data,
                   columns = ['product','price','qty']
                  )
ds1
```

```python
dss = ds1.sort_values(by = 'qty', ascending =False)
dss
```

```python
dss['TotalSales'] = dss['price']*dss['qty']
dss
```

```python
import pandas as pd
df = pd.read_csv('seoul.csv', encoding='cp949')
df
```

```python
df.columns = ['날짜', '지점','평균기온','최저기온','최고기온']
df
```

```python
df = df.drop(['지점'],axis=1)
df
```

```python
df['최고기온'].max()
```

```python
# 8 일교차
df['일교차'] = df['최고기온'] - df['최저기온']
dfg = df[df['일교차'] == df['일교차'].max()]['날짜']
dfm = df['일교차'].max()
print('일교차가 가장큰날 :', dfg, ', 일교차 :', dfm)
```

```python
df.mean()
```

```python
# 10 일교차
df['날짜'] = pd.to_datetime(df['날짜'])
df['월'] = df['날짜'].dt.month
# df
df.groupby('월').mean()['일교차']
```
